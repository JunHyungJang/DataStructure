```python
# importing required modules for this assignment

import random
from time import time_ns
from matplotlib import pyplot as plt
import math
```

```python
# Generate a list of N random numbers

def gen_list(N):
    L = []

    for i in range(N):
        L.append(i)
        
    random.shuffle(L)    
    return L

print(gen_list(10))
```

## Problem 1

**adding one item at the end of a list**

`L.append(random number)`

```python
# Write your codes for the analysis here
problem_size = []
result = []

for i in range(10000,100000,1000):
    L = gen_list(i)
    k = random.randint(0,i-1)
    start = time_ns()
    L.append(k)
    end = time_ns()
    elapsed = end - start
    
    problem_size.append(i)
    result.append(elapsed)

plt.figure(figsize = (10,10))
plt.scatter(problem_size, result)
    
    
```

## Problem 2

**adding one item at the middle of a list**

`L.insert(random position, random number)`

```python
# Write your codes for the analysis here
problem_size = []
result = []

for i in range(10000,100000,1000):
    L = gen_list(i)
    k = random.randint(0,i-1)
    t = random.randint(0,i-1)
    start = time_ns()
    L.insert(k,t)
    end = time_ns()
    elapsed = end - start
    
    problem_size.append(i)
    result.append(elapsed)

print(problem_size)
print(result)
plt.figure(figsize = (10,10))
plt.scatter(problem_size, result)
    
    
```

## Problem 3

**sorting a list**

`L.sort()`

```python
# Write your codes for the analysis here
problem_size = []
result = []

for i in range(10000,100000,1000):
    L = gen_list(i)
    start = time_ns()
    L.sort()
    end = time_ns()
    elapsed = end - start
    
    problem_size.append(i)
    result.append(elapsed)

print(problem_size)
print(result)
plt.figure(figsize = (10,10))
plt.scatter(problem_size, result)
```

## Problem 4

**find an item in a list**

`L[random number within the list] in L`

```python
# Write your codes for the analysis here
problem_size = []
result = []

for i in range(10000,100000,100):
    L = gen_list(i)
    k = random.randint(0,i-1)
    start = time_ns()
    t = L[k]
    for j in L:
        if j == t:
            break
    
    end = time_ns()
    elapsed = end - start
    
    problem_size.append(i)
    result.append(elapsed)

print(problem_size)
print(result)
plt.figure(figsize = (10,10))
plt.scatter(problem_size, result)
```

## Problem 5

**remove one item from a list**

`L.remove(L[random index])`

```python
# Write your codes for the analysis here
problem_size = []
result = []

for i in range(10000,100000,1000):
    L = gen_list(i)
    k = random.randint(0,i-1)
    start = time_ns()
    L.remove(k)
    end = time_ns()
    elapsed = end - start
    
    problem_size.append(i)
    result.append(elapsed)

print(problem_size)
print(result)
plt.figure(figsize = (10,10))
plt.scatter(problem_size, result)

```
